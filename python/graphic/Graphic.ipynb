```python
x = np.linspace(0, 2*np.pi, 500)

c = np.cos(x)

s = np.sin(x)


plt.figure("Gráficos Cossenoidal", figsize=(8, 4))

plt.subplots_adjust(
    left=0.15, 
    right=0.943,
    wspace=0.438,
    hspace= 0.736
)

axi1 = plt.subplot(2,1,1)
plt.plot(x, c)

axi1.set_xlabel('Eixo de Tempo')
axi1.set_ylabel('Eixo da amplitude')
axi1.grid()

axi1.set_title("Gráfico do Cosseno")


axi2 = plt.subplot(2,1,2)
plt.plot(x, s)
axi2.set_xlabel('Eixo de Tempo')
axi2.set_ylabel('Eixo da amplitude')
axi2.grid()

axi2.set_title("Gráfico do Seno")

plt.show()

```

```python
from matplotlib import pyplot as plt
import numpy as np

plt.style.use('ggplot')

x = np.linspace(1, 5, 500)
y = np.log10(x)

fig, axi = plt.subplots(figsize=(7, 4))

axi.plot(x, y, lw=1.2)

axi.plot([0,2.5],[0.4, 0.4], color="grey", linestyle = "--", lw=0.8)

#eixo y
axi.plot([2.5,2.5],[0, 0.4], color="grey", linestyle = "--", lw=0.8)

axi.plot(2.5, 0.4, marker="o", color="grey")

axi.set_xticks(np.arange(0, 5.5, 0.5))

plt.title("Gráfico do Cosseno", fontsize=16)
plt.xlabel("Eixo do x", fontsize=16)
plt.ylabel("Eixo y", fontsize=16)

plt.show
```

```python
from matplotlib import pyplot as plt
import numpy as np







plt.style.use('ggplot')

x = np.linspace(1, 5, 500)
y = np.log10(x)

fig, axe = plt.subplots(figsize=(7, 4))

axe.plot(x, y, lw=1.2)


axe.text(2.6, 0.35, "P(2,5; 0,4)")
axe.text(3, 0.42, "Logarítimos $y = log_{10}x$",
        fontsize=10,
        bbox=dict(facecolor="red", alpha=0.5))

axe.annotate("P(2,5; 0,4)", xy=(2.5, 0.4),
           fontsize=14, xytext=(0.5, 0.5),
           arrowprops=dict(facecolor="red"),
           color="r")


axe.plot([0, 2.5], [0.4, 0.4], 
         color="gray", linestyle = "--", lw=0.8)

#eixo y
axe.plot([2.5, 2.5], [0, 0.4], 
         color="gray", linestyle = "--", lw=0.8)

axe.plot(2.5, 0.4, marker="o", color="gray")

axe.set_xticks(np.arange(0, 5.5, 0.5))

plt.title("Gráfico do Cosseno", fontsize=16)
plt.xlabel("Eixo do x", fontsize=16)
plt.ylabel("Eixo y", fontsize=16)

plt.show
```
