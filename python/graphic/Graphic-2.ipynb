```python
import matplotlib.pyplot as plt


x = [1,2,3,4]

y =  [2,3,4,3]

plt.plot(x, y, label= 'dados')

plt.ylabel('Eixo y')

plt.xlabel('Eixo x')

plt.title('Ttulo do Grfico')

plt.show()
```
